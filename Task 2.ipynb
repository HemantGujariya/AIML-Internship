```python
import seaborn as sns
import matplotlib.pyplot as plt
titanic = sns.load_dataset('titanic')

# Histogram: Age
sns.histplot(titanic['age'].dropna(), bins=25, color='navy')
plt.title('Distribution of Passenger Age')
plt.show()

# Histogram: Fare
sns.histplot(titanic['fare'].dropna(), bins=30, color='teal')
plt.title('Distribution of Fare Paid')
plt.show()

# Boxplot: Age by Passenger Class
sns.boxplot(x='pclass', y='age', data=titanic)
plt.title('Age by Passenger Class')
plt.show()

# Boxplot: Age by Survival
sns.boxplot(x='survived', y='age', data=titanic)
plt.title('Age by Survival Status')
plt.show()
```
